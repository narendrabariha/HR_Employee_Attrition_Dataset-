# Outlier Detection and Removal Using IQR

```python
import pandas as pd
df = pd.read_csv("HR_Employee_Attrition_Dataset.csv")
df
```

```python
df.describe()
```

# Detect outliers using IQR

```python
Q1 = df.PercentSalaryHike.quantile(0.25)
Q3 = df.PercentSalaryHike.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

**Here are the outliers**

```python
df[(df.PercentSalaryHike<lower_limit)|(df.PercentSalaryHike>upper_limit)]
```

# Remove outliers

```python
df_no_outlier = df[(df.PercentSalaryHike>lower_limit)&(df.PercentSalaryHike<upper_limit)]
df_no_outlier
```
