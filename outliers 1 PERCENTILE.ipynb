```python
import pandas as pd
```

```python
df = pd.read_csv("HR_Employee_Attrition_Dataset.csv")
```

```python
df.head()
```

# Detect outliers using percentile

```python
max_thresold = df['YearsWithCurrManager'].quantile(0.95)
max_thresold
```

```python
df[df['YearsWithCurrManager']>max_thresold]
```

```python
min_thresold = df['YearsWithCurrManager'].quantile(0.05)
min_thresold
```

```python
df[df['YearsWithCurrManager']<min_thresold]
```

# Remove outliers

```python
df[(df['YearsWithCurrManager']<max_thresold) & (df['YearsWithCurrManager']>min_thresold)]
```
